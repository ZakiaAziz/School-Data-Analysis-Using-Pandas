```python
import pandas as pd
import numpy as np
```

```python
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
school_data.head()
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#SCHOOL SUMMARY

total_Schools = school_data_complete["school_name"].nunique()
total_students=school_data_complete["student_name"].count()
total_budget = school_data["budget"].sum()

average_maths=school_data_complete["maths_score"].mean()
average_reading=school_data_complete["reading_score"].mean()

maths_pass=school_data_complete.loc[student_data['maths_score']>=50]
percentage_maths=len(maths_pass)/total_students*100

reading_pass=school_data_complete.loc[student_data['reading_score']>=50]
percentage_reading=len(reading_pass)/total_students*100

overall_pass=school_data_complete.loc[(student_data['maths_score']>=50) & (student_data['reading_score']>=50)]
overall_percentage=len(overall_pass)/total_students*100


student_summary = pd.DataFrame({"Total Schools":15,"Total Students":total_students,"Total Budget":total_budget,"Average Maths Score":average_maths,"Average Reading Score":average_reading,"%Passing Maths":percentage_maths,"%Passing Reading":percentage_reading,"%Overall Passing":overall_percentage},index=[0])
student_summary["Total Budget"]=student_summary["Total Budget"].map("${:,.2f}".format)
student_summary
```

```python
#SCHOOL SUMMARY

school_types = school_data.set_index(["school_name"])["type"]
per_school_counts= school_data_complete["school_name"].value_counts()


per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_student_budget = per_school_budget/per_school_counts

per_school_maths = school_data_complete.groupby(["school_name"]).mean()["maths_score"]
per_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]


students_passing_maths=school_data_complete[(school_data_complete["maths_score"]>=50)]

students_passing_reading=school_data_complete[(school_data_complete["reading_score"]>=50)]
overall_passing= school_data_complete[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)]

per_school_maths_rate = students_passing_maths.groupby(["school_name"]).count()["student_name"]/per_school_counts*100

per_school_reading_rate = students_passing_reading.groupby(["school_name"]).count()["student_name"]/per_school_counts*100

overall_passing_rate = overall_passing.groupby(["school_name"]).count()["student_name"]/per_school_counts*100

school_summary=pd.DataFrame({"School Type": school_types,"Total Students":per_school_counts,"Total School Budget":per_school_budget,"Per Student Budget":per_student_budget,"Average Maths Score":per_school_maths,"Average Reading Score": per_school_reading,"% Passing Maths":per_school_maths_rate,"%Passing Reading":per_school_reading_rate,"% Overall Passing":overall_passing_rate})
#school_summary["Total School Budget"]=school_summary["Total School Budget"].map("${:,.2f}".format)
#school_summary["Per Student Budget"]=school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary
```

```python
#TOP PERFORMING SCHOOL

school_summary.sort_values(by='% Overall Passing',ascending=False,inplace=True)
top_schools=school_summary.head()
top_schools
```

```python
#BOTTOM PERFORMING SCHOOL

school_summary.sort_values(by='% Overall Passing',ascending=True,inplace=True)
bottom_schools=school_summary.head()
bottom_schools
```

```python
#MATHS SCORE BY YEAR

school_summary_spending=school_summary.reset_index(drop=True)


maths_score_9= school_data_complete.loc[student_data['year']==9]
maths_score_10=school_data_complete.loc[student_data['year']==10]
maths_score_11=school_data_complete.loc[student_data['year']==11]
maths_score_12=school_data_complete.loc[student_data['year']==12]

average_maths_9 = maths_score_9.groupby(["school_name"]).mean()["maths_score"]
average_maths_10 = maths_score_10.groupby(["school_name"]).mean()["maths_score"]
average_maths_11 = maths_score_11.groupby(["school_name"]).mean()["maths_score"]
average_maths_12=maths_score_12.groupby(["school_name"]).mean()["maths_score"]

Maths_Scores_By_Year=pd.DataFrame({"Year 9":average_maths_9, "Year 10":average_maths_10,"Year 11":average_maths_11,"Year 12":average_maths_12})
Maths_Scores_By_Year
```

```python
#READING SCORE BY YEAR

reading_score_9= school_data_complete.loc[student_data['year']==9]
reading_score_10=school_data_complete.loc[student_data['year']==10]
reading_score_11=school_data_complete.loc[student_data['year']==11]
reading_score_12=school_data_complete.loc[student_data['year']==12]

average_reading_9 = maths_score_9.groupby(["school_name"]).mean()["reading_score"]
average_reading_10 = maths_score_10.groupby(["school_name"]).mean()["reading_score"]
average_reading_11 = maths_score_11.groupby(["school_name"]).mean()["reading_score"]
average_reading_12=maths_score_12.groupby(["school_name"]).mean()["reading_score"]
Reading_Scores_By_Year=pd.DataFrame({"Year 9":average_reading_9, "Year 10":average_reading_10,"Year 11":average_reading_11,"Year 12":average_reading_12})
Reading_Scores_By_Year
```

```python
#SCORES BY SCHOOL SPENDING

spending_bins = [0, 585, 630, 645, 680]
group_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

pd.cut(school_summary_spending["Per Student Budget"],spending_bins,labels=group_labels, include_lowest=True)
school_summary_spending["Spending Ranges (Per Student)"]=pd.cut(school_summary_spending["Per Student Budget"],spending_bins,labels=group_labels, include_lowest=True)

spending_maths_scores = school_summary_spending.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"]
spending_reading_scores = school_summary_spending.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_maths = school_summary_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"]
spending_passing_reading = school_summary_spending.groupby(["Spending Ranges (Per Student)"]).mean()["%Passing Reading"]
overall_passing_spending = school_summary_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

Spending_summary=pd.DataFrame({"Average Maths Score": spending_maths_scores,"Average Reading Score":spending_reading_scores,"%Passing Maths":spending_passing_maths,"%Passing Reading":spending_passing_reading, "%Overall Passing":overall_passing_spending})
Spending_summary
```

```python
#SCORES BY SCHOOL SIZE

bins=[0,1000,2000,5000]
labels=["Small(<1000)","Medium(1000-2000)","Large(2000-5000)"]

pd.cut(school_summary_spending["Total Students"],bins,labels=labels, include_lowest=True)
school_summary_spending["School Size"]=pd.cut(school_summary_spending["Total Students"],bins,labels=labels, include_lowest=True)

size_maths_scores = school_summary_spending.groupby(["School Size"]).mean()["Average Maths Score"]
size_reading_scores = school_summary_spending.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_maths = school_summary_spending.groupby(["School Size"]).mean()["% Passing Maths"]
size_passing_reading = school_summary_spending.groupby(["School Size"]).mean()["%Passing Reading"]
overall_passing_bysize = school_summary_spending.groupby(["School Size"]).mean()["% Overall Passing"]

Size_summary=pd.DataFrame({"Average Maths Score":size_maths_scores ,"Average Reading Score":size_reading_scores,"%Passing Maths":size_passing_maths,"%Passing Reading":size_passing_reading, "%Overall Passing":overall_passing_bysize})
Size_summary
```

```python
#SCORES BY SCHOOL TYPE

Type_maths_scores = school_summary_spending.groupby(["School Type"]).mean()["Average Maths Score"]
Type_reading_scores = school_summary_spending.groupby(["School Type"]).mean()["Average Reading Score"]
Type_passing_maths = school_summary_spending.groupby(["School Type"]).mean()["% Passing Maths"]
Type_passing_reading = school_summary_spending.groupby(["School Type"]).mean()["%Passing Reading"]
overall_passing_bytype = school_summary_spending.groupby(["School Type"]).mean()["% Overall Passing"]

Type_summary=pd.DataFrame({"Average Maths Score":Type_maths_scores ,"Average Reading Score":Type_reading_scores,"%Passing Maths":Type_passing_maths,"%Passing Reading":Type_passing_reading, "%Overall Passing":overall_passing_bytype})
Type_summary
```

```python
#FORMATED DATAFRAME FOR BUDGET

school_summary=pd.DataFrame({"School Type": school_types,"Total Students":per_school_counts,"Total School Budget":per_school_budget,"Per Student Budget":per_student_budget,"Average Maths Score":per_school_maths,"Average Reading Score": per_school_reading,"% Passing Maths":per_school_maths_rate,"%Passing Reading":per_school_reading_rate,"% Overall Passing":overall_passing_rate})
school_summary["Total School Budget"]=school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"]=school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary
```
